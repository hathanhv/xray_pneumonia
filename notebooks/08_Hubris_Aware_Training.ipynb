# Hubris-Aware Boundary Training

Hard-label training, soft-label training, model selection, evaluation, and Hubris scoring now live in reusable source modules. This notebook only launches the YAML-driven pipeline.

```python
from pathlib import Path
import subprocess
import sys

project_root = Path.cwd()
if not (project_root / 'configs').exists():
    project_root = project_root.parent
config = project_root / 'configs/experiments/hubris_aware_boundary_training.yaml'
assert config.exists(), config
lung_app_python = Path.home() / '.conda/envs/lung_app/python.exe'
python_executable = lung_app_python if lung_app_python.exists() else Path(sys.executable)
print(f'Notebook kernel : {sys.executable}')
print(f'Training Python : {python_executable}')
print(f'Config          : {config}')
```

```python
command = [
    str(python_executable),
    str(project_root / 'scripts/train_hubris_aware_classifier.py'),
    '--config', str(config),
]
result = subprocess.run(
    command,
    cwd=project_root,
    env={**__import__('os').environ, 'MPLBACKEND': 'Agg'},
    text=True,
    stdout=subprocess.PIPE,
    stderr=subprocess.STDOUT,
)
print(result.stdout)
if result.returncode != 0:
    raise RuntimeError(f'Hubris training failed with exit code {result.returncode}')
```
